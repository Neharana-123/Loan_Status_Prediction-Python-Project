imbalanced-learn (or imblearn) is a Python library used to handle imbalanced datasets in machine learning. It provides tools to resample data, helping improve model performance when one class has significantly fewer samples than others.

```python
from imblearn.over_sampling import RandomOverSampler
```

The RandomOverSampler from imbalanced-learn is used to increase the number of samples in the minority class by randomly duplicating existing samples. This helps balance imbalanced datasets without creating synthetic data like SMOTE.

```python
oversample = RandomOverSampler(random_state=42)
x_resampled,y_resampled = oversample.fit_resample(x,y)

df_resampled=pd.concat([pd.DataFrame(x_resampled,columns= x.columns),pd.Series(y_resampled,name="Loan_status")],axis=1)
```

- random_state=42 ensures reproducibility, meaning the same results appear each time you run the code
- fit_resample(x, y) applies oversampling by duplicating random samples from the minority class (Loan_status = 0).
- Creates a new DataFrame with resampled features and target variable (Loan_status).
- ensures that Loan_status is correctly attached to the feature variables.
- After oversampling, both classes have equal instances.

- Creates a new DataFrame (df_resampled) with the updated data

```python
from sklearn.metrics import classification_report

def generate_classification_report(model_name, y_test, y_pred):
  report = classification_report(y_test,y_pred)
  print(f"Classification Report for {model_name}:\n{report}\n")

generate_classification_report(model1,y_resampled_test,y_pred_model1)
generate_classification_report(model2,y_resampled_test,y_pred_model2)
generate_classification_report(model3,y_resampled_test,y_pred_model3)
generate_classification_report(model4,y_resampled_test,y_pred_model4)
```

Each classification report includes:

Precision: How many predicted positives are actually positive?
Recall: How many actual positives were correctly predicted?
F1-score: The harmonic mean of precision and recall (balances both metrics).
Accuracy: Overall percentage of correct predictions.

#Key Observations
Balanced Data Improves Minority Class Recall

Before Balancing: Models predicted Loan Rejections (0s) poorly, leading to low recall (many false negatives).
After Balancing: Recall improved, meaning the models better recognized rejected loans.
Decision Tree and Random Forest Performed Better on Balanced Data

Decision Tree: Improved from 71% to 81% accuracy.

Random Forest: Best performer  Accuracy jumped from 79% to 88%.

Logistic Regression Performed Worse on Balanced Data--Accuracy dropped from 77% to 69%  This happens because oversampling increases variance, making logistic regression less effective.

KNN Performance Stayed Similar-- Accuracy remained around 71%-73%, meaning KNN isnt affected much by class imbalance.

# Conclusion & Best Model Choice
